#### Question 1: Find the questions and answers of last month.

```python
# your code here
SITE = StackAPI('stackoverflow')
```

```python
first_day = "01/05/2020"
today = '31/05/2020'
from_day = int(time.mktime(datetime.datetime.strptime(first_day, "%d/%m/%Y").timetuple()))
to_today = int(time.mktime(datetime.datetime.strptime(today, "%d/%m/%Y").timetuple()))
```

```python
questions = SITE.fetch('questions', fromdate = from_day, todate = to_today)
```

```python
questions500 = [question['link'].split('/')[5].replace('-', ' ') for question in questions['items']]
```

```python
len(questions500)
```

There is a limitation of 500 items per fetch in the API.

```python
answers = SITE.fetch('answers', fromdate = from_day, todate = to_today, filter='withbody')
```

```python
answers
```

There are more answer between those dates but there is a limitation of 5 pages per fetch by the API.
