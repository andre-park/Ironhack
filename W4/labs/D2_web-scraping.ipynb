#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
response = requests.get(url)
soup = BeautifulSoup(response.content)
````

````python
# your code here
soup.find_all('a', attrs = {'class':"link-box"})[0].find('strong').text
````

````python
import unicodedata
````

````python
tags = soup.find_all('a', attrs = {'class':"link-box"})
languages_articles = [tag.find("strong").text + ' ' + unicodedata.normalize('NFKD', tag.find('bdi').text) for tag\
                      in tags]
languages_articles
````

#### Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# your code here
infos = soup.find_all('td', attrs = {'class':"titleColumn"})
````

````python
infos
````

````python
infos[0].find('a')['title'].split(' (dir.)')[0]
````

````python
infos[0].find('a')['title'].split('), ')[1]
````
